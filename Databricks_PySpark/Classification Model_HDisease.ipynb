# Apache Spark: Classification Model

Using the ISLR dataset, I built a binary classification model to predict whether or not a patient has heart disease (`AHD`) given the following features:

- `Age`
- `Sex`
- `ChestPain`
- `RestBP`
- `Chol`
- `Fbs`
- `RestECG`
- `MaxHR`
- `ExAng`
- `Oldpeak`
- `Slope`
- `Ca`
- `Thal`

## Scoring

**Random Forest scored 85.6 Accuracy on training data, 82.6 on testing data**

```python
#Imports
from pyspark.ml.feature import StringIndexer, VectorAssembler, Tokenizer
from pyspark.ml.feature import OneHotEncoder, StopWordsRemover, CountVectorizer
from pyspark.ml import Pipeline
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.evaluation import BinaryClassificationEvaluator, MulticlassClassificationEvaluator
from pyspark.sql.types import IntegerType
from pyspark.ml.classification import RandomForestClassifier
```

```python
#Read CSV
df1 = spark.read.format("csv").option("header", "True").option('inferSchema', 'True').load("dbfs:/FileStore/tables/Heart-1.csv")

df1.show(5)
```

```python
#Show DTYPES
df1.printSchema()
```

## String Index Necessary Features

```python
#String Index Target y
labelindexer = StringIndexer(inputCol='AHD', outputCol='label')

df1 = labelindexer.fit(df1).transform(df1)

df1.show(5)
```

```python
#String Index Target x1
labelindexer_2 = StringIndexer(inputCol='ChestPain', outputCol='ChestPain_1')
df1 = labelindexer_2.fit(df1).transform(df1)
```

```python
#String Index Target x2
labelindexer_3 = StringIndexer(inputCol='Thal', outputCol='Thal_1')
df1 = labelindexer_3.fit(df1).transform(df1)
```

```python
#One Hot Encode X variables
encoder = OneHotEncoder(
    inputCols=["ChestPain_1", "Thal_1"],  
    outputCols=["ChestPain_OHE", "Thal_OHE"]
)

df1 = encoder.fit(df1).transform(df1)
```

```python
#Select Features
assembler = VectorAssembler(inputCols=['Sex', 'ChestPain_OHE', 'RestBP', 'Chol', 'Fbs', 
                                       'RestECG', 'MaxHR', 'ExAng', 'Oldpeak', 'Slope', 'Ca', 
                                       'Thal_OHE'
                                      ], outputCol='features')
```

```python
#Transform to new column
df1 = assembler.transform(df1)
df1.show(5)
```

## Train/Test Split

```python
#Set up Train/Test Split
train, test = df1.randomSplit([.8, .2], seed=42)
```

```python
train.count()
```

```python
test.count()
```

```python
#Instantiate Random Forest Classifier
rf = RandomForestClassifier(numTrees=50, maxDepth=3)
```

```python
#Fit to training data
rfModel = rf.fit(train)
```

```python
#train_pred = rfModel.transform(train)
```

```python
#test_pred = rfModel.transform(test)
```

```python
#Establish Multiclass Evaluator
evaluatorMulti = MulticlassClassificationEvaluator(labelCol="label")
```

```python
#classification code adapted from https://stackoverflow.com/questions/60772315/how-to-evaluate-a-classifier-with-pyspark-2-4-5

#Evaluate on Training Predictions

predictionAndTarget = rfModel.transform(train).select("label", "prediction")

acc = evaluatorMulti.evaluate(predictionAndTarget, {evaluatorMulti.metricName: "accuracy"})
f1 = evaluatorMulti.evaluate(predictionAndTarget, {evaluatorMulti.metricName: "f1"})
Precision = evaluatorMulti.evaluate(predictionAndTarget, {evaluatorMulti.metricName: "weightedPrecision"})
Recall = evaluatorMulti.evaluate(predictionAndTarget, {evaluatorMulti.metricName: "weightedRecall"})


```

```python
#Evaluate on Test Predictions

predictionAndTarget = rfModel.transform(test).select("label", "prediction")

acc1 = evaluatorMulti.evaluate(predictionAndTarget, {evaluatorMulti.metricName: "accuracy"})
f11 = evaluatorMulti.evaluate(predictionAndTarget, {evaluatorMulti.metricName: "f1"})
Precision1 = evaluatorMulti.evaluate(predictionAndTarget, {evaluatorMulti.metricName: "weightedPrecision"})
Recall1 = evaluatorMulti.evaluate(predictionAndTarget, {evaluatorMulti.metricName: "weightedRecall"})

```

```python
#Train Metrics
print("Accuracy:", acc)
print("F1 Score:", f1)
print("Weighted Precision:", Precision)
print("Weighted Recall:", Recall)
```

```python
#Test Metrics
print("Accuracy:", acc1)
print("F1 Score:", f11)
print("Weighted Precision:", Precision1)
print("Weighted Recall:", Recall1)
```
